```python
import matplotlib.pyplot as plt
%matplotlib inline

x= [1,2,3,4,5,6,7]
y=[50,51,52,49,48,43,52]
```

```python
plt.plot(x,y,color="red",linewidth=10,linestyle="dotted")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("graph")
plt.show()
```

```python
plt.plot(x,y,"g-")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("graph")
plt.show()
```

```python
plt.plot(x,y,"rD",alpha=0.5)
plt.show() #alpha transparency
```
