```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

```python
x=["google","amazon","microsoft","facebook"]
y=[90,136,89,45]
z=[45,65,23,22]

xpos = np.arange(len(x))
xpos
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.xticks(xpos,x)
plt.bar(xpos-0.2,y,color="blue",width=0.4,label="revenue")
plt.bar(xpos+0.2,z,color="orange",width=0.4,label="profit")
plt.xlabel("companies")
plt.ylabel("bllions")
plt.legend(loc="best")
plt.title("finance")
plt.show()
```
