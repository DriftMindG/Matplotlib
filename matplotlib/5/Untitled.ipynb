```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
blood_sugar_man = [113,85,90,150,140,88,93,115,135,80,77,82,129]
blood_sugar_women = [73,45,150,90,70,120,113,95,135,80,77,82,129]
```

```python
plt.hist(blood_sugar_man,bins=5, rwidth = 0.80 )
```

```python
plt.show()
```

```python
#can give legth also to filll the bins
plt.hist(blood_sugar_man,bins=[80,100,125,200],rwidth = 0.35,color="red" )
plt.show()
```

```python
#can give legth also to filll the bins
plt.hist([blood_sugar_man,blood_sugar_women],bins=[80,100,125,150],rwidth = 0.95,color=["orange","green"],label=["man","women"] )
plt.title("blood sugar analysis")
plt.xlabel("sugar ranges")
plt.ylabel("no of patients")
plt.legend()
plt.show()
```
