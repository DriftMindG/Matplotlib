```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = [1,2,3,4,5,6,7]
max_t =[56,24,33,42,53,62,77]
min_t = [25,15,22,23,44,34,56]
avg_t=[35,20,26,23,45,45,65]
```

```python
plt.plot(x,max_t,label= "max",color="red")
plt.plot(x,min_t,label="min",color="green")
plt.plot(x,avg_t,label="avg",color="yellow")
plt.xlabel("days")
plt.ylabel("temperature")
plt.title("temperature vs day graph")
plt.legend(loc="best")
plt.grid()
plt.show()
```
